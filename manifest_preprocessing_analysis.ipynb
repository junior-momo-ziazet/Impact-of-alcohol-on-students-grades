```python
import pandas as pd
```

```python
drebin = pd.read_csv("./datasets/drebin_manifest_analysis.csv")
androzoo = pd.read_csv("./datasets/androzoo_manifest_analysis.csv")
```

```python
df = pd.concat([drebin, androzoo])
```

```python
df = df.drop(["id"], axis=1)
df = df.sum(axis=0)
```

```python
empty_features = []
for i in range(len(df)):
    if df[i] == 0:
        empty_features.append(i)
```

```python
import matplotlib.pyplot as plt

df.plot()
plt.xticks(color='w')
for x in empty_features:
    plt.axvline(x=x, color="#ff000033")
plt.xlabel("Permissions")
plt.ylabel("Frequency")
plt.title("Permissions Frequency in the Manifest Dataset")
plt.show()
```

```python
print(len(df))
print(len(df) - len(empty_features))
```

```python
scores = [0.8723500225529995,
          0.8183029764549089,
          0.8256880733944953,
          0.8499324019828752,
          0.8601676224084693,
          0.603813559322034]

labels = [ "RF", "DT", "KNN", "MLP", "SVC", "NB"]
```
